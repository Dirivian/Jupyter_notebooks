# SVD Reconstruction

A look at singular value decomposition

```python
%pylab inline
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
```

Plotting a weird function

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
x= np.linspace(-10,10,100)
t=np.linspace(-10,10,100)
[T,X]=np.meshgrid(t,x)
Z=1/np.cosh(X)
P=Z*(1-0.5*np.cos(2*T))+(Z*np.tanh(X))*(1-0.5*np.sin(2*T));
# Plot the surface
ax.plot_surface(X,T,P, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.show()
```

SVD time

```python
U, s, V = np.linalg.svd(P, full_matrices=False)
```

Let's see the modes.

```python
plt.plot(s,'ko')
plt.show()
```

Plotting the percentages of the first ten modes

```python
plt.plot(s[0:10]*100/sum(s),'ko')
plt.show()
```

Only the first two are relevant. Let's see them

```python
plt.plot(U[:,0])
plt.show()
```

```python
plt.plot(U[:,1])
plt.show()
```

The Reconstruction

```python
def reconstruct(t,U,S,V):
    U1=np.array(U[:,0:t])
    i,j=np.indices(P.shape)
    s1=np.zeros([100,100])
    s1[i==j]=s
    s2=np.array(s1[0:t,0:t])
    V1=np.array(V[0:t,:])
    A1=np.matmul(s2,V1)
    A2=np.matmul(U1,A1)
    return A2
```

One mode reconstruction

```python
t=1
R1= reconstruct(t,U,s,V)
fig1 = plt.figure()
ax1 = fig1.gca(projection='3d')
# Plot the surface
ax1.plot_surface(X,T,R1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.show()
```

Two mode reconstruction

```python
t=2
R2= reconstruct(t,U,s,V)
fig2 = plt.figure()
ax2 = fig2.gca(projection='3d')
# Plot the surface
ax2.plot_surface(X,T,R2, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.show()
```
