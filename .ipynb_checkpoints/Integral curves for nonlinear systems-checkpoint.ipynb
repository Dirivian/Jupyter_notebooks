# Solving Riemann problems using Hugoniot loci and integral curves

We solve the following nonlinear system of equations.

$ v_t -u_x =0 $

$ u_t + p(v)_x =0$

```python
%matplotlib inline
from pylab import *
```

Where p is defined below

```python
p = lambda v: -exp(v)    
```

The two sets of hugoniot loci passing through the point are found using the function below.

```python
def hugoniot_loci(vstar,ustar):
    """
    ustar and vstar are a single state or a list of states.
    """
    if type(ustar) in [float,int]:
        # turn scalars into lists with one element:
        ustar = [ustar]
        vstar = [vstar]
        
    for vs,us in zip(vstar,ustar):
        # loop over each set of states
        v = linspace(-3,5,1000)
        s = sqrt(-(p(v)-p(vs))/(v-vs))
        uplus = us + s*(v-vs)
        uminus = us - s*(v-vs)
        plot(v,uplus,'r')
        plot(v,uminus,'b')
        plot(vs,us,'ko')
    legend(['1-waves','2-waves'], loc='lower left');
```

```python
vl,ul = 1,1
vr,ur= 4,3
hugoniot_loci([vl,vr], [ul,ur])
```

Similarly, for the integral curves,

```python
def integral_curve(vstar,ustar):
    """
    ustar and vstar are a single state or a list of states.
    """
    if type(ustar) in [float,int]:
        # turn scalars into lists with one element:
        ustar = [ustar]
        vstar = [vstar]
        
    for vs,us in zip(vstar,ustar):
        # loop over each set of states
        v = linspace(-3,5,1000)
        uplus = us - 2*(exp(vs/2)-exp(v/2))
        uminus = us + 2*(exp(vs/2)-exp(v/2))
        plot(v,uplus,'r')
        plot(v,uminus,'b')
        plot(vs,us,'ko')
    legend(['1-waves','2-waves'], loc='lower left');
```

```python
vl,ul = 1,1
vr,ur= 4,3
integral_curve([vl,vr], [ul,ur])
```

Finding the point of intersection of integral curves

```python
from scipy.optimize import fsolve  # for solving nonlinear equation
```

```python
def g(vm,vl,ul,vr,ur):

    phi1 = ul - 2*(exp(vl/2)-exp(vm/2))
    phi2 =  ur + 2*(exp(vr/2)-exp(vm/2))
    return phi1 - phi2
```

```python
fsolution = fsolve(g, x0=3, args=(vl,ul,vr,ur))
vm = fsolution[0]

um = ul - 2*(exp(vl/2)-exp(vm/2))
print('Two-rarefaction solution: vm = %g, um = %g' %(vm,um))
```

Using integral curves would mean that the middle state is connected to the left and right states by rarefactions. For the left state [1,1], there is only a certain range of right states which gives a 2-rarefaction solution. Thatr range is plotted below.

```python
integral_curve(1,1)
text(vl+0.2, ul, '$q_\ell$', fontsize=20)

for v in linspace(-3,5,40):
    uplus =  ul - 2*(exp(vl/2)-exp(v/2))
    uminus =ul + 2*(exp(vl/2)-exp(v/2))
    umin = min(uplus, uminus)
    plot([v,v],[-30,umin],'g-')
```

We see that [4,3] does not fall in that range. The actual solution for it, the one that doesn't violate entropy conditions is a shock -rarefaction. Using the intersection of the appropriate sets of integral and hugoniot curves, we find the exact solution.

```python
def integral_hugoniot_curve(vstar,ustar):
    """
    ustar and vstar are a single state or a list of states.
    """
    if type(ustar) in [float,int]:
        # turn scalars into lists with one element:
        ustar = [ustar]
        vstar = [vstar]
        
    for vs,us in zip(vstar,ustar):
        # loop over each set of states
        v = linspace(-3,5,1000)
        s = sqrt(-(p(v)-p(vs))/(v-vs))
        uplus = us + s*(v-vs)
        uminus =us + 2*(exp(vs/2)-exp(v/2))
        plot(v,uplus,'r')
        plot(v,uminus,'b')
        plot(vs,us,'ko')
    legend(['1-waves','2-waves'], loc='lower left');
```

```python
vl,ul = 1,1
vr,ur= 4,3
text(vl+0.2, ul, '$q_\ell$', fontsize=20)
text(vr+0.2, ur, '$q_r$', fontsize=20)
integral_hugoniot_curve([vl,vr], [ul,ur])
```

```python
def ginthugo(vk,vl,ul,vr,ur):
    s = sqrt(-(p(vl)-p(vk))/(vl-vk))
    s1 = sqrt((exp(vk)-exp(1))/(vk-1))
    phi1 = 1 + s1*(vk-1)
    phi2 =  3 + 2.*(exp(4.0/2.0)-exp(vk/2.))
    return phi1 - phi2
```

```python
vl,ul = 1,1
vr,ur= 4,3
fsolution = fsolve(ginthugo, x0=3, args=(vl,ul,vr,ur))
vk = fsolution[0]

uk = ur + 2*(exp(vr/2)-exp(vk/2))
print('The correct solution: vm = %g, um = %g' %(vk,uk))
```

```python
vl,ul = 1,1
vr,ur= 4,3
text(vl+0.2, ul, '$q_\ell$', fontsize=20)
text(vr+0.2, ur, '$q_r$', fontsize=20)
text(vm+0.2, um, '$q_m$', fontsize=20)

integral_hugoniot_curve([vl,vr], [ul,ur])
plot(vm,um,'ko')
```
